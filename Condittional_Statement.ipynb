### **1. Introduction**
- What are conditional statements?
- Importance of conditional logic in programming.
- Overview of conditional statements in Python.

---

### **2. if Statement**
- Syntax:
```python
if condition:
    # code block
```
- Example:
```python
x = 10
if x > 5:
    print("x is greater than 5")
```

---

### **3. if-else Statement**
- Syntax:
```python
if condition:
    # code block if condition is True
else:
    # code block if condition is False
```
- Example:
```python
age = 18
if age >= 18:
    print("Eligible to vote")
else:
    print("Not eligible to vote")
```

---

### **4. if-elif-else Statement**
- Syntax:
```python
if condition1:
    # code block for condition1
elif condition2:
    # code block for condition2
else:
    # code block if all conditions are False
```
- Example:
```python
marks = 85
if marks >= 90:
    print("Grade: A")
elif marks >= 75:
    print("Grade: B")
else:
    print("Grade: C")
```

---

### **5. Nested if Statements**
- Syntax:
```python
if condition1:
    if condition2:
        # code block if both conditions are True
```
- Example:
```python
x = 10
y = 20
if x > 5:
    if y > 15:
        print("x and y satisfy the conditions")
```

---

### **6. Logical Operators in Conditions**
- `and`, `or`, `not`
- Example:
```python
x = 10
y = 20
if x > 5 and y > 15:
    print("Both conditions are True")
```

---

### **7. Comparison Operators**
- Examples of using `<, >, <=, >=, ==, !=`
```python
x = 10
if x == 10:
    print("x is equal to 10")
```

---

### **8. Ternary Operator**
- Syntax:
```python
value = true_expression if condition else false_expression
```
- Example:
```python
age = 18
status = "Adult" if age >= 18 else "Minor"
print(status)
```

---

### **9. Using Membership Operators in Conditions**
- `in`, `not in`
- Example:
```python
fruits = ["apple", "banana", "cherry"]
if "apple" in fruits:
    print("Apple is in the list")
```

### **10. Examples and Exercises**
#### Example 1: Check if a number is positive, negative, or zero.
```python
number = 0
if number > 0:
    print("Positive")
elif number < 0:
    print("Negative")
else:
    print("Zero")
```

#### Example 2: Check if a year is a leap year.
```python
year = 2024
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("Leap year")
else:
    print("Not a leap year")
```

#### Example 3: Password verification.
```python
password = "secure123"
if len(password) >= 8:
    print("Password is valid")
else:
    print("Password must be at least 8 characters long")
```
